```python
with open("Inputs/Day12.txt", "r") as input:
    input = input.read()

input_list = input.split("\n")
```

```python
broken_springs_list = []
broken_springs_count = []

for line in input_list:
    broken_springs_list.append(line.split(" ")[0])
    broken_springs_count.append(line.split(" ")[1].split(","))
```

```python
broken_springs_count
```

```python
part1_answer = 0
for lineN in range(0, len(broken_springs_list)):
# for lineN in range(0, 1):
    # print(lineN)
    num_unknowns = len(broken_springs_list[lineN]) - len(broken_springs_list[lineN].replace("?", ""))
    # print(num_unknowns)
    for i in range(0, 2**num_unknowns):
        unknowns = '{0:0{num_unknowns}b}'.format(i, num_unknowns=num_unknowns)
        unknowns = unknowns.replace("0", ".").replace("1", "#")
        new_string = ""
        num_unknowns_count = 0
        for index, letter in enumerate(broken_springs_list[lineN]):
            if letter == "?":
                new_string += unknowns[num_unknowns_count]
                num_unknowns_count += 1
            else:
                new_string += letter
        new_string = new_string.split(".")
        print(new_string)
        new_string_2 = []
        for part in new_string:
            if part != "":
                new_string_2.append(part)
        print(new_string_2)
        string_works = True
        if len(new_string_2) == len(broken_springs_count[lineN]):
            for partN in range(0, len(new_string_2)):
                if len(new_string_2[partN]) != int(broken_springs_count[lineN][partN]):
                    # print(len(new_string_2[partN]))
                    # print(broken_springs_count[lineN][partN])
                    string_works = False
        else:
            string_works = False
        if string_works:
            part1_answer += 1
        # print(part1_answer)
```
