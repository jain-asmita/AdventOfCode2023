```python
input_grid = []
for line in input_list:
    input_grid.append([*line])
```

```python
maxRow = len(input_grid) - 1
maxColumn = len(input_grid[0]) - 1
```

```python
def should_continue(currentRow, currentColumn, iHaveBrokenDownHelp):
    if iHaveBrokenDownHelp:
        return False
    if input_grid[currentRow][currentColumn] == "S":
        return False
    if input_grid[currentRow][currentColumn] == ".":
        return False
    if currentRow > maxRow:
        return False
    if currentRow < 0:
        return False
    if currentColumn > maxColumn:
        return False
    if currentColumn < 0:
        return False
    return True
    
```

```python
currentRow = startRow - 1
currentColumn = startColumn
```

```python
def solveGrid(lastRow, lastColumn, currentRow, currentColumn, stepsList):
    iHaveBrokenDownHelp = False
    while(should_continue(currentRow, currentColumn, iHaveBrokenDownHelp)):
        stepsList.append(tuple([currentRow, currentColumn]))
        if input_grid[currentRow][currentColumn] == "|":
            if (currentRow - lastRow) == 1 and currentColumn == lastColumn:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow + 1
                currentColumn = currentColumn
            elif (currentRow - lastRow) == -1 and currentColumn == lastColumn:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow - 1
                currentColumn = currentColumn
            else:
                iHaveBrokenDownHelp = True
        elif input_grid[currentRow][currentColumn] == "-":
            if (currentColumn - lastColumn) == 1 and currentRow == lastRow:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow
                currentColumn = currentColumn + 1
            elif (currentColumn - lastColumn) == -1 and currentRow == lastRow:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow
                currentColumn = currentColumn - 1
            else:
                iHaveBrokenDownHelp = True
        elif input_grid[currentRow][currentColumn] == "L":
            if (currentRow - lastRow) == 1 and currentColumn == lastColumn:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow
                currentColumn = currentColumn + 1
            elif (currentColumn - lastColumn) == -1 and currentRow == lastRow:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow - 1
                currentColumn = currentColumn
            else:
                iHaveBrokenDownHelp = True
        elif input_grid[currentRow][currentColumn] == "J":
            if (currentRow - lastRow) == 1 and currentColumn == lastColumn:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow
                currentColumn = currentColumn - 1
            elif (currentColumn - lastColumn) == 1 and currentRow == lastRow:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow - 1
                currentColumn = currentColumn
            else:
                iHaveBrokenDownHelp = True
        elif input_grid[currentRow][currentColumn] == "7":
            if (currentRow - lastRow) == -1 and currentColumn == lastColumn:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow
                currentColumn = currentColumn - 1
            elif (currentColumn - lastColumn) == 1 and currentRow == lastRow:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow + 1
                currentColumn = currentColumn
            else:
                iHaveBrokenDownHelp = True
        elif input_grid[currentRow][currentColumn] == "F":
            if (currentRow - lastRow) == -1 and currentColumn == lastColumn:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow
                currentColumn = currentColumn + 1
            elif (currentColumn - lastColumn) == -1 and currentRow == lastRow:
                lastRow = currentRow
                lastColumn = currentColumn
                currentRow = currentRow + 1
                currentColumn = currentColumn
            else:
                iHaveBrokenDownHelp = True
    stepsList.append(tuple([currentRow, currentColumn]))
    return stepsList
```

```python
up = solveGrid(startRow, startColumn, startRow-1, startColumn, [tuple([startRow, startColumn])])
if len(up) > 1:
    if input_grid[up[-1][0]][up[-1][1]] == "S":
        real_loop = up
down = solveGrid(startRow, startColumn, startRow+1, startColumn, [tuple([startRow, startColumn])])
if len(down) > 1:
    if input_grid[down[-1][0]][down[-1][1]] == "S":
        real_loop = down
left = solveGrid(startRow, startColumn, startRow, startColumn - 1, [tuple([startRow, startColumn])])
if len(left) > 1:
    if input_grid[left[-1][0]][left[-1][1]] == "S":
        real_loop = left
right = solveGrid(startRow, startColumn, startRow, startColumn + 1, [tuple([startRow, startColumn])])
if len(right) > 1:
    if input_grid[right[-1][0]][right[-1][1]] == "S":
        real_loop = right
print(len(real_loop) // 2)
```

```python
covered_cells = []
for newRowN in range(0, (maxRow+1)*3):
    for newColN in range(0, (maxColumn+1)*3):
        if new_grid[newRowN][newColN] == "#":
            covered_cells.append(tuple([newRowN, newColN]))
```

```python
def hasReachedEdge(tile):
   return tile[0] < 0 or tile[0] > 3*(maxRow+1) or tile[1] < 0 or tile[1] > 3*(maxColumn+1)
```

```python
def isTileAccessible(tile, covered_cells, visited, known_accessible_tiles):
    increment_row_tile = list(tile)
    increment_row_tile[0] += 1
    increment_row_tile = tuple(increment_row_tile)
    if increment_row_tile in known_accessible_tiles:
        return True, covered_cells
    elif not hasReachedEdge(increment_row_tile):
        if increment_row_tile not in visited:
            if increment_row_tile not in covered_cells:
                visited.add(increment_row_tile)
                if isTileAccessible(increment_row_tile, covered_cells, visited, known_accessible_tiles):
                    return True, covered_cells
                else:
                    covered_cells.append(increment_row_tile)
    else:
        return True, covered_cells

    increment_column_tile = list(tile)
    increment_column_tile[1] += 1
    increment_column_tile = tuple(increment_column_tile)
    if increment_column_tile in known_accessible_tiles:
        return True, covered_cells
    elif not hasReachedEdge(increment_column_tile):
        if increment_column_tile not in visited:
            if increment_column_tile not in covered_cells:
                visited.add(increment_column_tile)
                if isTileAccessible(increment_column_tile, covered_cells, visited, known_accessible_tiles):
                    return True, covered_cells
                else:
                    covered_cells.append(increment_column_tile)
    else:
        return True, covered_cells

    decrement_row_tile = list(tile)
    decrement_row_tile[0] -= 1
    decrement_row_tile = tuple(increment_row_tile)
    if decrement_row_tile in known_accessible_tiles:
        return True, covered_cells
    elif not hasReachedEdge(decrement_row_tile):
        if decrement_row_tile not in visited:
            if decrement_row_tile not in covered_cells:
                visited.add(decrement_row_tile)
                if isTileAccessible(decrement_row_tile, covered_cells, visited, known_accessible_tiles):
                    return True, covered_cells
                else:
                    covered_cells.append(decrement_row_tile)
    else:
        return True, covered_cells
    
    decrement_column_tile = list(tile)
    decrement_column_tile[1] -= 1
    decrement_column_tile = tuple(decrement_column_tile)
    if decrement_column_tile in known_accessible_tiles:
        return True, covered_cells
    elif not hasReachedEdge(decrement_column_tile):
        if decrement_column_tile not in visited:
            if decrement_column_tile not in covered_cells:
                visited.add(decrement_column_tile)
                if isTileAccessible(decrement_column_tile, covered_cells, visited, known_accessible_tiles):
                    return True, covered_cells
                else:
                    covered_cells.append(decrement_column_tile)
    else:
        return True, covered_cells
    return False, covered_cells
```

```python
num_accessible_tiles = 0
known_accessible_tiles = set()
for newRowN in range(0, 3*(maxRow+1)):
    for newColN in range(0, 3*(maxColumn+1)):
        print(f'{newRowN} {newColN}')
        tile = tuple([newRowN, newColN])
        if tile not in pipe_cells:
            isAccessible, covered_cells = isTileAccessible(tile, covered_cells, set(), known_accessible_tiles)
            if isAccessible:
                num_accessible_tiles += 1
                known_accessible_tiles.add(tile)
            else:
                covered_cells.add(tile)
print(((3*(maxRow+1)) * (3*(maxColumn+1)) - len(pipe_cells) - num_accessible_tiles)/ 9)
```
