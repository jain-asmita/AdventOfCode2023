```python
import re
```

```python
nums_list = []
for line in input_list:
    nums_line_list = line.split(" ")
    new_list = []
    for num in nums_line_list:
        new_list.append(int(num))
    nums_list.append(new_list)
nums_list
```

```python
def areAllZeros(num_list):
    # print(num_list)
    # if len(num_list) == 1:
    #     return True
    for num in num_list:
        if num != 0:
            return False
    return True
    # print(num_list)
    # return num_list[-1] == 0
```

```python
def complete_differences_lists(differences_lists):
    next_num = 0
    for i in range(len(differences_lists) - 1, -1, -1):
        next_num = next_num + differences_lists[i][-1]
    return next_num
        
```

```python
def subtractLists(list1, list2):
    list3 = []
    for i in range(0, len(list1)):
        list3.append(list1[i] - list2[i])
    return list3
```

```python
def findNextNum(num_list):
    differences_lists = []
    differences_lists.append(num_list)
    i = 0
    while(not areAllZeros(differences_lists[i])):
        differences_lists.append(subtractLists(differences_lists[i][1:], differences_lists[i][:-1]))
        i+=1
    
    # print(differences_lists)
    return complete_differences_lists(differences_lists)
    
```

```python
part_1_answer = 0
for num_list in nums_list:
    num_list.reverse()
    part_1_answer += findNextNum(num_list)
```
